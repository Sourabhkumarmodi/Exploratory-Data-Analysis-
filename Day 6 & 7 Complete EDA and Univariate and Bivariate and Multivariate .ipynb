```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing Dataset 1: Titanic

```python
titanic=pd.read_csv("D://titanic.csv")
```

```python
titanic.head()
```

# 2<Numerical-Categorical> | Bar plot

```python
sns.barplot(x='Pclass',y='Age',data=titanic)
```

```python
sns.barplot(x='Pclass',y='Age',data=titanic,hue='Sex')
```

```python
sns.barplot(x='Pclass',y='Fare',data=titanic,hue='Sex')
```

# 3<Categorical-Numerical> |Boxplot

```python
sns.boxplot(x='Sex',y='Age',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',data=titanic,hue='Survived')
```

# 4<Numerical-Categorical>|Distplot

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# 5<Categorical-Categorical>|Heatmap

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

# 6<Categorical-Categorical>|ClusterMap

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
